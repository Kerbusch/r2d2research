```python
import numpy as np
import os
import io
import PIL
import PIL.Image
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.layers import *
import numpy as np 
import cv2 
import os
```

```python
#Load the dataset with all the labeld graphs 
train_dataset = tf.keras.utils.image_dataset_from_directory("/home/wilhelm/Documents/GitHub/r2d2research/sua/Dataset/train",
                                      batch_size= 10, shuffle = True, image_size=(200,200))
```

```python
#Load the dataset that is used for validation
validation_dataset = tf.keras.utils.image_dataset_from_directory("/home/wilhelm/Documents/GitHub/r2d2research/sua/Dataset/Test",
                                                          batch_size=50, image_size=(200,200))
```

```python
#check which classes are incluced in the dataset
train_dataset.class_names
class_names = train_dataset.class_names
print(class_names)
```

```python
#Data augmentation used to flip, rotate and zoom the data in the dataset, by using this we can emulate different 
#graphs instead of collecting it.
data_augmentation = tf.keras.Sequential(
  [
    tf.keras.layers.RandomFlip("horizontal",input_shape=(200,200,3)),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomRotation(0.2)
  ]
)

#saves the best model with val_accuracy
best_callbacks = [tf.keras.callbacks.EarlyStopping(patience = 1000) ,
                 tf.keras.callbacks.ModelCheckpoint(filepath="/home/wilhelm/Documents/GitHub/r2d2research/sua",monitor="val_accuracy",
                    save_best_only=True,
                    save_weights_only=False,
                    mode="auto",
                    save_freq="epoch",)]
```

```python
#this is convolutional block class, it is ues to recreate a Residual block. it looks at differnce between y and x 
#and uses the output.

# most important formula = R(x) = (uitgang - ingang) = O(x)-x
class ResidualBlock(tf.keras.Model):
    def __init__(self, num_filters, kernel_size=(3, 3)):
        super(ConvBlock, self).__init__()
        self.conv = Conv2D(num_filters, kernel_size)
        self.bn = BatchNormalization()
        self.elu = Activation("elu")
        self.pooling = MaxPool2D((2, 2))

    def call(self, x, pool=True):
        y = self.conv(x)
        x = self.bn(y)
        x = self.elu(x)
        x = Add()([x,y])
        x = self.elu(x)
        if pool == True:
            x = self.pooling(x)
            
        return x
#the build model is made with 4 convolutional blocks( residual blocks ) and after that a global average pooling
#layer, a dropout layer (which temporaly removes some of the data per epoch) and finally a dense layer with 
#a softmax function.
def build_model(input_shape):
    inputs = Input(input_shape)
    inputs = data_augmentation(inputs)
    x = ResidualBlock(32)(inputs)
    x = ResidualBlock(64)(x)
    x = ResidualBlock(128)(x, pool=False)
    x = ResidualBlock(128)(x, pool=False)
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.2)(x)
    x = Dense(6, activation="softmax")(x)
    
    model = tf.keras.models.Model(inputs, x)
    return model
```

```python
#input shape is 200 by 200
model = build_model((200,200, 3))
```

```python
#comiler setting with a focus on the accuracy
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=(0.001),
                                                 beta_1=(0.9),
                                                 beta_2=(0.999),
                                                 epsilon=(1e-7),),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])
```

```python
#start of training the model with all the settings mentioned before
model.fit = model.fit(train_dataset,epochs=150, 
                      validation_data=validation_dataset, 
                      shuffle=True,
                      callbacks=best_callbacks )
```
